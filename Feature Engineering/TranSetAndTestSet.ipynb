### <font color = yellow>Checking Outlier

```python
df['Number_of_Reviewer'] = pd.to_numeric(df['Number_of_Reviewer'], errors='coerce')
print(df['Number_of_Reviewer'].min())
print(df['Number_of_Reviewer'].max())
```

```python
Q1 = df['Number_of_Reviewer'].quantile(0.25)
Q2 = df['Number_of_Reviewer'].median()
Q3 = df['Number_of_Reviewer'].quantile(0.75)
print([Q1,Q2,Q3])
```

```python
IQR = Q3-Q1
IQR
```

```python
outlier_set = df[(df['Number_of_Reviewer'] < (Q1 - 1.5 * IQR)) | (df['Number_of_Reviewer'] > (Q3 + 1.5 * IQR))]
outlier_set
```

```python
pd.DataFrame(outlier_set).to_csv('Outlier.csv')
```

### <font color = yellow>Train set & Test Set

<b><font color = yellow>A test set</font></b> is a subset of data used to evaluate the performance of a model that has already been trained on a training set. It is important to have a separate test set to ensure that the model is not overfitting to the training data and is able to generalize well to new, unseen data.

Typically, the test set is randomly sampled from the original dataset, with a portion of the data reserved for testing and the rest used for training. The size of the test set can vary depending on the size of the original dataset, but a common practice is to use 20% of the data for testing and 80% for training.

Once the model has been trained on the training set, it is evaluated on the test set to determine its performance. The test set can also be used for parameter tuning or model selection, but care must be taken to avoid overfitting to the test data.

<b><font color = yellow>A training set</font></b> is a dataset used to train a model. It is a subset of the total dataset and is used to build the model's parameters. The remaining portion of the dataset, called the validation set, is used to evaluate the performance of the model during training, while the test set is used to evaluate the final performance of the model after it has been trained. 

Therefore, to define the training set, you need to select a subset of the available data that you will use to train your machine learning model. Typically, this involves randomly selecting a portion of the data, such as 70% or 80%, and setting it aside for training. The remaining data can be split further into a validation set and a test set. The validation set is used to monitor the performance of the model during training, while the test set is only used at the end to evaluate the final performance of the trained model. 

It is important to note that the selection of the training, validation, and test sets should be done carefully to ensure that they are representative of the overall data distribution and that the model's performance on these subsets accurately reflects its performance on new, unseen data.

```python
from sklearn.model_selection import train_test_split

# Assume the data is stored in a pandas DataFrame called 'data'
x = df.drop('Rating', axis=1) # Input features (all columns except the 'Rating' column)
y = df['Rating'] # Output variable (the 'Rating' column)

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```
